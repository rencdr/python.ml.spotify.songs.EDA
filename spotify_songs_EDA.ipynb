```python
playlist_counts = ds['playlist_genre'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(playlist_counts.index, playlist_counts.values)
plt.xlabel('Playlist Genre')
plt.ylabel('Song Count')
plt.title('Number of Songs by Playlist Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
playlist_counts = ds['playlist_genre'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(playlist_counts.values, labels=playlist_counts.index, autopct='%1.1f%%')
plt.title('Percentage of Songs by Playlist Genre')
plt.show()
```

```python
release_dates = ds['track_album_release_date'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(release_dates.index, release_dates.values)
plt.xlabel('Release Date')
plt.ylabel('Song Count')
plt.title('Number of Songs by Release Date')
plt.xticks(rotation=45)
plt.show()
```

```python
ds[['playlist_genre', 'playlist_subgenre']] = ds[['playlist_genre', 'playlist_subgenre']] \
                                                  .apply(lambda x: x.str.capitalize(), axis=1)
fig = px.sunburst(ds,
                  path=['playlist_genre', 'playlist_subgenre'],
                  color='track_popularity',
                  labels={'track_popularity': 'Popularity'})
fig.show()
```

```python
fig = px.treemap(ds.groupby(['track_artist', 'track_name']).track_popularity.mean().sort_values(ascending=False).reset_index()[:100],
                 path=['track_artist', 'track_name'],
                 values='track_popularity',
                 title="Artists"
                )
fig.show()
```

```python
fig = px.treemap(ds.groupby(['playlist_genre', 'playlist_subgenre']).track_popularity.agg(['count', 'mean']).reset_index(),
                 path=[px.Constant('Playlist Genres'), 'playlist_genre', 'playlist_subgenre'],
                 values='count',
                 color='mean',
                 labels={'mean': 'Popularity'},
                )
fig.show()
```
